```python
#identifiyng garbage values
for i in data.select_dtypes(include="object").columns:
    print(data[i].value_counts())
    print("***"*10)
```

```python
# Finding missing value percentages
missing_data = data.isnull().mean() * 100
missing_data = missing_data[missing_data > 0]
missing_data.sort_values(ascending=False, inplace=True)

# Visualising missing value percentages
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_data.index, y=missing_data)
plt.xticks(rotation=90)
plt.title("Missing Value Percentages in Columns")
plt.ylabel("Missing Value (%)")
plt.show()
```

```python
# 'sold' sütununda sınıf dağılımını inceleyelim (1: Satıldı, 0: Satılmadı)
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='window_items_sold')
plt.title("Satılma Durumunun Dağılımı")
plt.xlabel("Satılma Durumu (1: Satıldı, 0: Satılmadı)")
plt.ylabel("Adet")
plt.show()
```

```python
# Check unique values in the target variable 'window_items_sold'
print(data['window_items_sold'].unique())

# Check the count of each unique value
print(data['window_items_sold'].value_counts())

# Create a new binary column for sold status (1 if sold, 0 if not sold)
data['window_items_sold_binary'] = data['window_items_sold'].apply(lambda x: 1 if x > 0 else 0)

# Plot the distribution of the new binary column
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='window_items_sold_binary')
plt.title("Distribution of Sold Status")
plt.xlabel("Sold Status (1: Sold, 0: Not Sold)")
plt.ylabel("Count")
plt.show()
```

```python
#histogram to understand distrubition
for i in data.select_dtypes(include="number").columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x=i, kde=True)
    plt.title(f"Distribution of {i}")
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#Boxplot to identify outliers
for i in data.select_dtypes(include="number").columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x=i)
    plt.title(f"Boxplot of {i}")
    plt.xlabel(i)
    plt.show()
```

```python
numeric_data = data.select_dtypes(include='number')
correlation = numeric_data.corr()['window_items_sold']

print(correlation.sort_values(ascending=True))
```

```python
# Select numeric data
numeric_data = data.select_dtypes(include='number')

# Calculate the correlation matrix
correlation = numeric_data.corr()

# Extract and sort the correlations with window_items_sold
correlation_window_items = correlation['window_items_sold'].sort_values(ascending=True)

# Visualizing correlations
plt.figure(figsize=(10, 6))
sns.barplot(x=correlation_window_items.index, y=correlation_window_items.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Correlation Values: window_items_sold')
plt.xlabel('Columns')
plt.ylabel('Correlation Value')
plt.show()
```
